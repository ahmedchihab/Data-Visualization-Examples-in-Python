```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.set_style("whitegrid")

data_normal = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(6,4))
sns.histplot(data_normal, kde=True, color="skyblue", bins=30)
plt.title("Normal Distribution\nSymmetrical: mean = median = mode")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
data_right = np.random.exponential(scale=1, size=1000)

plt.figure(figsize=(6,4))
sns.histplot(data_right, kde=True, color="orange", bins=30)
plt.title("Right-Skewed (Positive)\nTail to the right, mean > median")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
data_left = -np.random.exponential(scale=1, size=1000)

plt.figure(figsize=(6,4))
sns.histplot(data_left, kde=True, color="purple", bins=30)
plt.title("Left-Skewed (Negative)\nTail to the left, mean < median")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
data_uniform = np.random.uniform(low=0, high=1, size=1000)

plt.figure(figsize=(6,4))
sns.histplot(data_uniform, kde=False, color="green", bins=30)
plt.title("Uniform Distribution\nAll values equally likely")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
data_bimodal = np.concatenate([np.random.normal(-2, 0.5, 500),
                               np.random.normal(2, 0.5, 500)])

plt.figure(figsize=(6,4))
sns.histplot(data_bimodal, kde=True, color="red", bins=30)
plt.title("Bimodal Distribution\nTwo peaks")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,2))
sns.boxplot(x=data_right, color="lightblue")
plt.title("Boxplot\nMedian, Quartiles, Outliers")
plt.show()
```

```python
x_pos = np.linspace(0, 10, 100)
y_pos = x_pos * 2 + np.random.normal(0, 2, 100)

plt.figure(figsize=(6,4))
sns.scatterplot(x=x_pos, y=y_pos, color="blue")
plt.title("Scatterplot: Positive Correlation")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x_neg = np.linspace(0, 10, 100)
y_neg = -x_neg * 2 + np.random.normal(0, 2, 100)

plt.figure(figsize=(6,4))
sns.scatterplot(x=x_neg, y=y_neg, color="red")
plt.title("Scatterplot: Negative Correlation")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x_none = np.linspace(0, 10, 100)
y_none = np.random.normal(0, 5, 100)

plt.figure(figsize=(6,4))
sns.scatterplot(x=x_none, y=y_none, color="green")
plt.title("Scatterplot: No Correlation")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
data_matrix = np.random.rand(6, 6)
mask = np.triu(np.ones_like(data_matrix, dtype=bool))

plt.figure(figsize=(6,4))
sns.heatmap(data_matrix, annot=True, fmt=".2f", cmap="coolwarm", mask=mask)
plt.title("Heatmap: Correlation Matrix")
plt.show()
```

```python
categories = np.random.choice(["A", "B", "C"], size=200)

plt.figure(figsize=(6,4))
sns.countplot(x=categories, hue=categories, palette="pastel", legend=False)
plt.title("Count Plot: Frequency of Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
sizes = [30, 45, 25]
labels = ["Group A", "Group B", "Group C"]

plt.figure(figsize=(5,5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=["#66b3ff","#99ff99","#ff9999"])
plt.title("Pie Chart")
plt.show()
```
